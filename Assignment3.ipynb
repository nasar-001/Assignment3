# ***NumPy***

```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)
zeros_array = np.zeros(5)
ones_array = np.ones(5)
full_array = np.full(5, "a")
print("Zeros:", zeros_array)
print("Ones:", ones_array)
print("Full:", full_array)
range_array = np.arange(0, 10, 2)
linspace_array = np.linspace(0, 1, 5)
print("Range Array:", range_array)
print("Linspace Array:", linspace_array)
```

# ***MatplotLib***

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
plt.plot(df['mpg'], df['cylinders'], marker='o', linestyle='-')
plt.title('mpg Vs Cylinders')
plt.xlabel('mpg')
plt.ylabel('cylinders')
plt.grid(True)
plt.show()
```

```python
grouped_salary = df.groupby('HP')['acceleration'].mean()
plt.figure(figsize=(6, 4))
plt.bar(grouped_salary.index, grouped_salary.values)
plt.title('Horse power with Acceleration')
plt.xlabel('Horse power')
plt.ylabel('Acceleration')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['year'], bins=5, edgecolor='black')
plt.title('Year Distribution')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(df['displacement'], df['weight'])
plt.title('Scatter Plot of Displacement vs. Weight')
plt.xlabel('Disp')
plt.ylabel('Weight')
plt.show()
```
